```python
import pandas as pd
import keras
import numpy as np
from sklearn import preprocessing
```

```python
data=pd.read_csv('drive/My Drive/water_potability.csv')
```

```python
#parsing data 
X1 = pd.DataFrame(data, columns= ['ph'])
X2 = pd.DataFrame(data, columns= ['Hardness'])
X3 = pd.DataFrame(data, columns= ['Solids'])
X4 = pd.DataFrame(data, columns= ['Chloramines'])
X5 = pd.DataFrame(data, columns= ['Sulfate'])
X6 = pd.DataFrame(data, columns= ['Conductivity'])
X7 = pd.DataFrame(data, columns= ['Organic_carbon'])
X8 = pd.DataFrame(data, columns= ['Trihalomethanes'])
X9 = pd.DataFrame(data, columns= ['Turbidity'])
Y = pd.DataFrame(data, columns= ['Potability'])
X1
```

```python
#Visual array dummy
datas = np.array([[1,1],[1,0],[0,1],[0,0]])
test = np.array([[1],[0],[0],[0]])
print(datas)
print(test)
```

```python
#Set Data Done
XS = np.array([X1,X2,X3,X4,X5,X6,X7,X8,X9],dtype=int)
XS = XS.transpose()
YS = np.array(Y,dtype=int)
#YS = Y.values.tolist()
XSS = XS[0]
YSS = YS[0]

print(XSS[0])
print(YS)
```

```python

#MODEL dummy
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[9],activation=keras.activations.tanh)])

#Model compile
model.compile(optimizer='adam',loss='mean_squared_error')
```

```python
#Arsitektur 
model.summary()
```

```python
#Bobot
weights = model.get_weights()
weights
```

```python
#train dummy
model.fit(XSS,YS,epochs=1000)
```

```python
#test data dummy
cek_data = [-9223372036854775808,204,20791,7,368,564,10,86,2]
#cek_data = [1,0]
jawaban = model.predict([cek_data])
jawaban
```
